```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("covid_deaths.csv")
df=pd.DataFrame(data)
df.head()
```

```python
df1=df[["Date Of Death Year","Date Of Death Month","AgeGroup","NaturalCause"]]
df1
```

```python
df1.isna()
```

```python
df2=df1[(df1["AgeGroup"])=="15-24 years"]
df2
```

```python
df3=df2.groupby(["Date Of Death Year","Date Of Death Month",]).sum()
df3
```

```python
palette=["red","yellow","green"]
fig=plt.figure(figsize=(12,6))
sns.lineplot(data=df3,x="Date Of Death Month",y="NaturalCause",hue="Date Of Death Year",palette=palette)
plt.xlabel("Month",fontweight="bold",fontsize=12)
plt.ylabel("Natural Cause deaths",fontweight="bold",fontsize=12)
plt.title("Natural Cause Deaths in US: 15-24y",fontweight="bold",fontsize=18)
plt.legend(title="Date Of Death Year")
plt.show()
```

```python
df4=df[["Date Of Death Year","Date Of Death Month","AgeGroup","NaturalCause","AllCause"]]
df4
```

```python
df4=df4[(df4["AgeGroup"])=="35-44 years"]
df4
```

```python
df4=df4.groupby(["Date Of Death Year"]).sum()
df4
```

```python
df4=df4[["NaturalCause","AllCause"]]
df4
```

```python
df4.plot(kind="bar",stacked=True)
plt.xlabel("Years",fontweight="bold",fontsize=12)
plt.ylabel("Excess Deaths",fontweight="bold",fontsize=12)
plt.title("Excess Deaths in US: 35-44 y",fontweight="bold",fontsize=18)
plt.legend(["Non-Covid","Covid"])
plt.show()
```

```python
df5=df[["Date Of Death Year","Date Of Death Month","AgeGroup","NaturalCause","AllCause"]]
df5

```

```python
df6=df5[(df5["AgeGroup"])=="15-24 years"]
df6
```

```python
df7=df6.groupby(["Date Of Death Year","Date Of Death Month"],as_index=False).sum()
df7
```

```python
df7["ExcessDeaths"]=df7["AllCause"]-df7["NaturalCause"]
df7
```

```python
fig=plt.figure(figsize=(18,9))
sns.barplot(data=df7,x="Date Of Death Month",y="ExcessDeaths",hue="Date Of Death Year",palette="flare")
plt.xlabel("Month",fontweight="bold",fontsize=12)
plt.ylabel("Excess deaths",fontweight="bold",fontsize=12)
plt.title("Monthly Excess Deaths in US: 15-24y",fontweight="bold",fontsize=18)
plt.legend(title="Date Of Death Year")
plt.show()
```
